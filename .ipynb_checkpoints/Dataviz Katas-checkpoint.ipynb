```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Graph 1

```python
x = np.linspace(0,50)
y = x /3
plt.plot(x, y, color="red", label='x=3y')
plt.title('Tracer une droite')
plt.xlabel('Axe des x')
plt.ylabel('Axe des y')
plt.ylim(0,20) 
plt.xlim(0,50)
plt.legend(loc='upper left')
plt.show()
```

### Graph 6

#### Import csv files

```python
transactions=pd.read_csv('transactions.csv')
products=pd.read_csv('products.csv')
customers=pd.read_csv('customers.csv')
```

```python
products['categ'].value_counts().plot.pie(autopct='%1.0f%%', title='Products category distribution')
```

### Graph 7

```python
transactions['date']=pd.to_datetime(transactions['date'].str.replace('test_', ''))
transactions = pd.merge(transactions, products, on='id_prod', how='left')
```

```python
transactions['month_year'] = pd.to_datetime(transactions['date']).dt.to_period('M').astype(str)
```

```python
sales_per_category= transactions.groupby(['month_year', 'categ'])['price'].sum()
```

```python
sales_per_month_by_categ= sales_per_category.unstack(level="categ")
```

```python
sales_per_month_by_categ.plot(use_index=True, title='Evolution of sales per month by category', xlabel = "month_year", ylabel="euros").legend(loc='upper left')
```
